```python
# Dependencies
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

# Url

```python
#url = 'https://fbref.com/en/comps/9/stats/Premier-League-Stats#stats_standard'
url = 'https://www.myfootballfacts.com/stats/premier-league-by-season/premier-league-2020-21/premier-league-combined-goals-assists-2020-21/'
```

```python
# Retrieve page with the requests module
response = requests.get(url)
```

```python
# Create BeautifulSoup object; parse with 'html.parser'
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
# Examine the results, then determine element that contains sought info
print(soup.prettify())
```

```python
# results are returned as an iterable list
results = soup.find_all('div', class_='wptm_table tablesorter-bootstrap')
print(results)
```

```python
r = requests.get(url)
df_list = pd.read_html(r.text) # this parses all the tables in webpages to a list
df = df_list[0]
df.head(2000)
```

```python
players = df[["Player", "Nationality"]]
players
```

```python
players.to_csv('Player_nationality.csv')
```
